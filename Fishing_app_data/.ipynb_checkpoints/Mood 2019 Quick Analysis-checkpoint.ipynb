# Form Dataframe

```python
data = pd.read_csv('data_dropped.csv', sep = ",", header = None)
```

```python
n = 0

while n < len(data):
    temp = pd.DataFrame({"1": data.iloc[n], "2": data.iloc[n+1]})
    temp = temp.T
    temp.columns = temp.iloc[0]
    temp = temp.iloc[1:]
    temp = temp.apply(pd.to_numeric, errors='ignore')
    u = temp['userKey'][0]
    
    env = temp['tripValue'].iloc[:,0][0]
    n = n + 2
    
    for index in range(temp.shape[1]):
        c = temp.iloc[: , index]
        name = temp.columns[index]
        ## hosuekeeping info ##
        if name == 'tripValue':
            env = c.values[0]
       
        # fish
        if name == 'fishTapLocation':
            userKey.append(u)
            tripValue.append(env)
            [r.append(np.nan) for r in rating]
            fishTapLocation.append(c.values[0])     
        if name == 'fishTapTime':
            fishTapTime.append(c.values[0])
        if name == 'fishPoints':
            fishPoints.append(c.values[0])
        if name == "fishReleaseTime":
            fishReleaseTime.append(c.values[0])
        
        # rating
        if name == 'ratingStartValue':
            userKey.append(u)
            tripValue.append(env)
            [f.append(np.nan) for f in fish]
            ratingStartValue.append(c.values[0])     
        if name == 'ratingStartTapTime':
            ratingStartTapTime.append(c.values[0])
        if name == 'ratingEndValue':
            ratingEndValue.append(c.values[0])
            username.append(u)
            endvalue.append(c.values[0])
        if name == 'ratingEndTapTime':
            ratingEndTapTime.append(c.values[0])
        if name == 'ratingTimeChange':
            ratingTimeChange.append(c.values[0])
    
    # standardize my list length
    length = [len(l) for l in all_lst]
    max_length = max(length)
    [f.extend((max_length - len(f))*[np.nan]) for f in fish]
    [r.extend((max_length - len(r))*[np.nan]) for r in rating]
    tripValue.extend((max_length - len(tripValue))*[np.nan])
    userKey.extend((max_length - len(userKey))*[np.nan])
        
        

```

```python
df = {'user':userKey, 'env':tripValue, 'price': fishPoints, 'tap': fishTapTime, 'release': fishReleaseTime, 'loc_f': fishTapLocation,
                        'rating_init': ratingStartValue, 'rating_end':ratingEndValue}

```
