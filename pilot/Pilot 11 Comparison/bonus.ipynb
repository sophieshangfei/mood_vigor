```python
# import file
data_1 = pd.read_csv("data_exp_27347-v3_task-d7fe.csv", sep = ",")
data_2 = pd.read_csv("data_exp_27347-v3_task-yaom.csv", sep = ",")
```

```python
data_1 = data_1[['Participant Public ID', 'Participant Private ID', 'UTC Timestamp','key pressed', 'price_displayed', 'environment', 'task', 'function id',
       'builtin rt', 'end_time', 'start_time', 'response', 'round_instance']]

data_2 = data_2[['Participant Public ID', 'Participant Private ID', 'UTC Timestamp','key pressed', 'price_displayed', 'environment', 'task', 'function id',
       'builtin rt', 'end_time', 'start_time', 'response', 'round_instance']]
```

```python
data = pd.concat([data_1, data_2])
```

```python
data = data.replace('null', np.nan)
data  = data.replace('nan', np.nan)
```

```python
data['start_time'] = round(data['start_time'].astype(float))
```

```python

data.rename(columns={"Participant Private ID": "Subject"}, inplace = True)
data.rename(columns={"Participant Public ID": 'external_id'}, inplace = True)
data.rename(columns={"UTC Timestamp": 'date'}, inplace = True)
```

```python
data=data[(data.task == "fishing_pos")|(data.task == "fishing_neg")|(data.task == "fishing_control")]
```

```python
data.columns
```

```python
bonus = data.sort_values('date').groupby(["external_id","environment", "price_displayed", "task"]).size()
bonus = bonus.reset_index()
bonus.columns = ['id',"enviornment", 'price_displayed', 'factor', 'tap']
bonus = bonus.replace('fishing_control', 4/3000)
bonus = bonus.replace('fishing_neg', 4/3000)
bonus = bonus.replace('rod_fixing', 0)
bonus.head(10)
```

```python
bonus = bonus.round(20)
bonus['earning'] = bonus['price_displayed']*bonus['factor']*bonus['tap']
pay = bonus.groupby(['id', "enviornment"])['earning'].sum().reset_index()
pay['earning'] = np.ceil(pay['earning'])
pay['prolific'] = [','.join(i) for i in zip(pay["id"].map(str),pay["earning"].astype(str))]
pay = pay[pay.enviornment == "post_mood"]
pay
```
